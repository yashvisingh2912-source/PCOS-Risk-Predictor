```python
import matplotlib.pyplot as plt

pcos_rate = data.groupby('Weight gain(Y/N)')['PCOS (Y/N)'].mean()
plt.figure(figsize=(3,2))
pcos_rate.plot(kind='bar')
plt.title('PCOS Rate by Weight gain(Y/N)')
plt.ylabel('PCOS Rate')
plt.show()
```

```python
pcos_rate = data.groupby('hair growth(Y/N)')['PCOS (Y/N)'].mean()
plt.figure(figsize=(3,2))
pcos_rate.plot(kind='bar')
plt.title('PCOS Rate by hair growth(Y/N)')
plt.ylabel('PCOS Rate')
plt.show()
```

```python
pcos_rate = data.groupby('Pregnant(Y/N)')['PCOS (Y/N)'].mean()
plt.figure(figsize=(3,2))
pcos_rate.plot(kind='bar')
plt.title('PCOS Rate by Pregnant(Y/N)')
plt.ylabel('PCOS Rate')
plt.show()
```

```python
fig, axes = plt.subplots(2, 4, figsize=(15, 8))
axes = axes.flatten()

binary_cols = ['Weight gain(Y/N)', 'hair growth(Y/N)', 'Skin darkening (Y/N)', 
               'Hair loss(Y/N)', 'Pimples(Y/N)', 'Fast food (Y/N)', 'Reg.Exercise(Y/N)']

for i, col in enumerate(binary_cols):
    pcos_rate = data.groupby(col)['PCOS (Y/N)'].mean()
    pcos_rate.plot(kind='bar', ax=axes[i], title=col)
    axes[i].set_ylabel('PCOS Rate')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.hist(data[data['PCOS (Y/N)']==0][' Age (yrs)'], alpha=0.5, label='No PCOS', bins=20)
plt.hist(data[data['PCOS (Y/N)']==1][' Age (yrs)'], alpha=0.5, label='PCOS', bins=20)
plt.legend()
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution by PCOS')
plt.show()
```
